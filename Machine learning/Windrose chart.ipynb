```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
df = pd.read_pickle("data/df_clean.pkl")
df.head()
```

```python
df["velocidade do vento a 10m"].dtypes
```

```python
df = df[df["direção do vento"] > 0]
```

```python
import windrose_labels as wi
ax = wi.WindroseAxes.from_ax()
ax.bar(df["direção do vento"], df["velocidade do vento a 10m"], normed=True, opening=1, edgecolor='white')


vals = ax.get_yticks()
ax.set_yticklabels(['{:7.2f}%'.format(x) for x in vals])
ax.set_rmax(25)


ax.set_legend(labels = ["(0 - 3.5)", '(3.5 - 6.5)', '(6.5 - 9)','(9 - 12)', '(12 - 15)'], title="Wind speed (m/s)", loc="lower left")
plt.title('Windrose',fontweight='bold')
```
