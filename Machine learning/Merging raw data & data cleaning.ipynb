It seems that the column of speed has been read erroneously, after seeing it with depth you can see that values ​​that are above 10 meters per second are altered by 1000, so I apply the following condition to divide by a thousand when they are above 10

```python

df2004["velocidade do vento a 10m"]  = np.where(df2004["velocidade do vento a 10m"]  < 10,
                                                df2004["velocidade do vento a 10m"],
                                                df2004["velocidade do vento a 10m"]/1000)
```

# 2006 

```python
df2006 = pd.read_excel("data/jaguaruana 2006 1-365.xlsx")
df2006.head()
```

```python
df2006["velocidade do vento a 10m"]  = np.where(df2006["velocidade do vento a 10m"]  < 10,
                                                df2006["velocidade do vento a 10m"],
                                                df2006["velocidade do vento a 10m"]/1000)

df_2006_columnas = df2006[['ano', 'dia', 'hora', 'temperatura interna',
                           'pressão hpa','velocidade do vento a 10m',
                           'direção do vento','Temperatura a 2m °C', 'umidade relativa a 2 m']]
df_2006_columnas.head()
```

# 2007

```python
df2007 = pd.read_excel("data/jaguaruana 2007 1-365.xlsx")
df2007.head()
```

```python
df2007["velocidade do vento a 10m"]  = np.where(df2007["velocidade do vento a 10m"]  < 10,
                                                df2007["velocidade do vento a 10m"],
                                                df2007["velocidade do vento a 10m"]/1000)

df_2007_columnas = df2007[['ano', 'dia', 'hora', 'temperatura interna',
                           'pressão hpa','velocidade do vento a 10m',
                           'direção do vento']]
df_2007_columnas.head()
```
