# Loading wind data

```python
df = pd.read_csv("../data/Vientodef.csv" , usecols =["Data","Direcao","Velocidade"])
df["Data"] = pd.to_datetime(df["Data"])
df["Ano"]= df["Data"].dt.year
df["Mes"]= df["Data"].dt.month
df["hour"]= df["Data"].dt.hour

df.head()
```

```python
velocidademean = df['Velocidade'].mean()
velocidademean
```

```python
velarray = df.sort_values(by=['Velocidade'])
velarray = velarray["Velocidade"]
```

# Weibull distribution

```python
import scipy.special
from scipy import stats

plt.xlabel('Velocidade do vento (m/s)', fontweight='bold')
plt.ylabel('Densidade de probabilidade', fontweight='bold')
plt.plot(velarray, stats.exponweib.pdf(velarray, *stats.exponweib.fit(velarray, 1, 1, scale=1, loc=0)), color='red',linestyle ="--", label = "Weibull")
_ = plt.hist(velarray, bins=np.linspace(0, 16, 12), normed=True, alpha=0.5, edgecolor = 'black',  linewidth=1,facecolor='Deepskyblue',label = "Dados empiricos");
plt.legend()
plt.show()
```

# Weibull parameters

```python
k3,c3 = stats.weibull_min.fit(velarray, loc=0)[0],stats.weibull_min.fit(velarray, loc=0)[2]
k3,c3
```

# Weibull cumulative distribution

```python

def weiba(x,c,k):
    return  1- (np.exp(-(x / c)**k))

plt.plot (velarray, weiba(velarray, c3, k3), color = "red",linestyle ="--", label = "Weibull")
    
_= plt.hist(velarray, bins=14, normed=True,cumulative=True, alpha=0.5, edgecolor = 'black',  linewidth=1, color ="Deepskyblue",label="Dados empiricos")   

plt.legend(loc='upper left')
plt.xlabel('Wind speed (m/s)', fontweight='bold')
plt.ylabel('CDF', fontweight='bold')
```

#  Wind characteristics table

```python
grouped_Ano = df.groupby("Ano")
import scipy.special
from scipy import stats
Massaar10 = 1.225-(0.000194*10)
Massaar10
```

```python
tabla = pd.DataFrame(
    {
        "Mean": grouped_Ano["Velocidade"].mean(),
        "SD" : grouped_Ano["Velocidade"].std(),
        "k (-)": grouped_Ano["Velocidade"].apply (lambda x: ((np.std(x)/np.mean(x))**(-1.086))),
        "C (m/s)": grouped_Ano["Velocidade"].apply (lambda x: (np.mean(x)*(1/scipy.special.gamma(1 + (1/((np.std(x)/np.mean(x))**(-1.086))))))),
        "WPD (W/m$^2$)": grouped_Ano["Velocidade"].apply (lambda x: 0.5*Massaar10*(((np.mean(x)*(1/scipy.special.gamma(1 + (1/((np.std(x)/np.mean(x))**(-1.086)))))))**3)*scipy.special.gamma(1 + (3/((np.std(x)/np.mean(x))**(-1.086)))))
        

    }
)
resultados = pd.DataFrame(tabla.mean(), columns=["Total_Overage"]).T
Tabla_final = pd.concat([
    tabla,
    resultados
], axis=0)
Tabla_final
```

# Wind turbine production

```python
h1 = 10
hr = 10
vh = 67
z0 = 0.02 #aeropuertos segun justus 
arrayc2 = []
arrayk2 = []
arrayCF = []
array = []
for i in aerogeneradores.index:
    alpha = (z0/hr)**0.2
    nnum = 1-(np.log(c)/np.log(vh))
    nden = 1-((alpha*(np.log(h1/hr)))/np.log(vh))
    n = alpha*(nnum/nden)
    c2 =  c3*(((aerogeneradores["Hub height (m)"][i])/h1)**n)
    arrayc2.append(c2)
    
    #aerogeneradores["c2"][i] = c*(((aerogeneradores["Hub height (m)"][i])/h1)**n)
    knum = 1-((alpha*np.log(h1/hr))/(np.log(vh)))
    kden = 1-((alpha*np.log((aerogeneradores["Hub height (m)"][i])/hr))/(np.log(vh)))
    k2 = k3*((knum)/(kden))
    arrayk2.append(k2)
    cfnum = (np.exp(-(aerogeneradores["Cut-in speed (m/s)"][i]/c2)**(k2)) - (np.exp(-(aerogeneradores["Rated speed (m/s)"][i]/c2)**k2)))
    cfden = ((aerogeneradores["Rated speed (m/s)"][i]/c2)**k2) - ((aerogeneradores["Cut-in speed (m/s)"][i]/c2)**(k2))
    cfrest = np.exp(-(aerogeneradores["Cut-out speed (m/s)"][i]/c2)**k2)
    CF = (cfnum/cfden)-cfrest
    arrayCF.append(CF)
aerogeneradores["c2"] = arrayc2
aerogeneradores["k2"] = arrayk2
aerogeneradores["Capacity factor"] = arrayCF
aerogeneradores["CF (%)"] = aerogeneradores["CF"]*100
aerogeneradores["Annual energy production(MWh)"] = (aerogeneradores["CF"]*aerogeneradores["Rated power(KW)"]*8760)/1000
tabla_potencia = (aerogeneradores[["Wind Turbine Model","c2","k2","Capacity factor","Annual energy production(MWh)"]])
```

```python
tabla_potencia = tabla_potencia.round(2)
tabla_potencia
```
